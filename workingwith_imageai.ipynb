```python
from imageai.Prediction import ImagePrediction
```

```python
import os
execution_path=os.getcwd()
```

```python
prediction= ImagePrediction()
```

```python
prediction.setModelTypeAsResNet50()
prediction.setModelPath(os.path.join(execution_path, "resnet50_imagenet_tf.2.0.h5"))
```

```python
prediction.loadModel()
```

```python
predictions, probabilities = prediction.predictImage(os.path.join(execution_path, "icecream1.png"), result_count=5 )
for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)
```

```python
predictions, probabilities = prediction.predictImage(os.path.join(execution_path, "racingcar1.jpeg"), result_count=5 )
for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)
```
